## Extracting stock data using webscraping
Using BeautifulSoup to scrap the stock data from websites of companies like Amazon and Netflix

```python
#Installing and importing libraries
#!pip install bs4
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
#Storing datasets' website url in variable
tesla_url="https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
gme_url="https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
```

```python
#Sending get request using requests.get and converting the fetched data to text
tesla_data=requests.get(tesla_url).text
gme_data=requests.get(gme_url).text
```

```python
#Now we will convert the text to html5 format using BeautifulSoup
tesla_soup=BeautifulSoup(tesla_data,'html5lib')
gme_soup=BeautifulSoup(gme_data,'html5lib')
```

### To convert html into pandas dataframe 
#### Method 1: Using read_html

```python
#Using BeautifulSoup or the read_html function extract the table with Tesla Quarterly Revenue and store it into a dataframe named tesla_revenue. The dataframe should have columns Date and Revenue.
read_html_pandas=pd.read_html(tesla_url)
tesla_revenue=read_html_pandas[1]

new_cols=["Date","Revenue"]
tesla_revenue.columns=new_cols
tesla_revenue.head()
```

```python
#Execute the following line to remove the comma and dollar sign from the `Revenue` column.
tesla_revenue["Revenue"] = tesla_revenue['Revenue'].str.replace(',|\$',"")
#Execute the following lines to remove an null or empty strings in the Revenue column.
tesla_revenue.dropna(inplace=True)

tesla_revenue = tesla_revenue[tesla_revenue['Revenue'] != ""]
```

```python
#Display the last 5 row of the `tesla_revenue` dataframe using the `tail` function. Take a screenshot of the results.

tesla_revenue.tail()
```

#### Method 2: Using find_all

```python
#Using beautiful soup extract the table with GameStop Quarterly Revenue and store it into a dataframe named gme_revenue. 
#The dataframe should have columns Date and Revenue. Make sure the comma and dollar sign is removed from the Revenue column 
gme_revenue = pd.DataFrame(columns = ['Date', 'Revenue'])

for row in gme_soup.find_all("tbody")[1].find_all("tr"):
    col = row.find_all("td")
    date = col[0].text
    revenue = col[1].text.replace("$", "").replace(",", "")
    
    gme_revenue = gme_revenue.append({"Date": date, "Revenue": revenue}, ignore_index = True)
```

```python
#Display the last five rows of the gme_revenue dataframe using the tail function. Take a screenshot of the results.\
gme_revenue.tail()
```
